En bedrift produserer goder med følgende produktfunksjon:


\begin{equation}
    x = f(N,K)=N^{0.5}K^{0.5}\tag{1}

På kort sikt har bedriften tilgang til begrenset kapital $K_0=10$. Prisen per enhet kapital er fast $r=500$ NOK, mens én time med arbeidskraft koster $w=350$ NOK. På kort sikt kan bedriften velge hvor mange timer arbeidskraft som tilsettes. Bedriften leier et bygg til produksjonen av produktet, og dette koster 10 000 NOK (faste kostnader).      
a) Finn et utrykk for bedriftens totale produksjonskostnad på kort sikt som en funksjon av produsert mengde og kapitalbeholdningen: $C_s(x, K_0)$. Tegn denne funksjon i en figur med produksjon på den horisontale aksen. Tegn også kostnadsfunksjonen som oppstår når kapitalbeholdningen økes til $K_1=50$. Kommenter dine funn.   
b) I en ny figur tegn totale gjennomsnittskostnader på kort sikt, variabler gjennomsnittskostnader på kort sikt, og grensekostnaden på kort sikt. Kommenter og forklar forholdet mellom total og variabel gjennomsnittskostnad og grensekostnaden.       
c) Hvilke skalaegenskaper viser denne produktfunksjon?

**Oppgave A**

```python
# Definerer produktformelen:
eq_prod = sp.Eq(N**0.5 * K**0.5, x)
eq_prod
```

```python
# Løser produktformelen mhp. N.
eq_prod_sol = sp.solve(eq_prod, N)
eq_prod_sol = eq_prod_sol[0]
eq_prod_sol
```

```python
# Transformerer fra sympy til lambda funksjon.
eq_func_lamb = sp.lambdify((x, K), eq_prod_sol)
```

```python
# Definerer kostnadsfunksjonen.
def cost(x, K):
    return 350*eq_func_lamb(x, K) + 500*K + 10e3
```

```python
# Lager figuren.
fig, ax = plt.subplots()
ax.set(xlim = (0, 100))
ax.set(ylim = (0, 350e3))
ax.set_title("Produksjonskostnader på kort sikt")
ax.set_ylabel("Kostnad (K)", loc = "top")
ax.set_xlabel("Kvantum (x)", loc = "right")

# Plotter inn grafene.
ax.plot(cost(x_01, 10), label = "Kapitalbeholdning = 10", color = "blue")
ax.plot(cost(x_01, 50), label = "Kapitalbeholdning = 50", color = "orange")

# Legger til en legend og grid.
ax.grid()
ax.legend(bbox_to_anchor = (1, 0.3))
```

```python
# Finner likevektskvantum (Ekstra :-)).
equilibrium = sp.Eq(cost(x, 10), cost(x, 50))
equilibrium_solved = sp.solve(equilibrium, x)
eq = round(equilibrium_solved[1], 2)
eq
```

På figuren over kan vi se produksjonkostnadene på kort sikt med en kapitalbeholdning på henholdsvise $K = 10$ og $K = 50$.\
Når $K = 50$ er det høyere kostnader med lavere produksjonsnivå og lavere kostnader med høyere produksjonsnivå. Dette kan man tolke som at med lav produksjon får man ikke utnyttet all produksjonen.\
Når $K = 10$ er det lavere kostnader med lavere produksjonsnivå og høyere kostnader med høyere produksjonsnivå. Dette kan man tolke som at med lav produksjon får man utnyttet produksjonen bedre enn med høyere realkapital.\
Vi kan altså se at det lønner seg med en lavere realkapital fram til **26.73** produserte enheter. 

**Oppgave B**

```python
# Definerer funksjonen for totale gj.snitt kostnader.
def total_cost(x, K):
    return cost(x,K)/x

# Definerer funksjonen for variable gj.snitt kostnader.
def variable_cost(x, K):
    return (cost(x, K)-10e3)/x

# Finner grensekostnaden.
marginal_cost_01 = sp.diff(cost(x, K), x)
marginal_cost_02 = sp.lambdify((x, K), marginal_cost_01) 
```

```python
# Lager figuren.
fig, ax = plt.subplots()
ax.set(xlim = (0, 40))
ax.set(ylim = (0, 4e3))
ax.set_title("Gj.snitt- og grensekostnad")
ax.set_ylabel("Kostnad (K)", loc = "top")
ax.set_xlabel("Kvantum (x)", loc = "right")

# Plotter inn grafene.
ax.plot(total_cost(x_01, 10), label = "Total gj.snittkostnader (K = 10)")
ax.plot(variable_cost(x_01, 10), label = "Variable gj.snittkostnader (K = 10)")
ax.plot(marginal_cost_02(x_01, 10), label = "Grensekostnad (K = 10)")

# Legger til en legend og grid.
ax.grid()
ax.legend(bbox_to_anchor = (1, 0.3))
```

På figuren over kan vi se totale gj.snittkostnader, variable gj.snittkostnader og grensekostnadene der $K = 10$.\
Når grensekostnaden skjærer den totale- og variable gj.snittkostnaden er gjennomsnittkostnaden lavest.\
Altså tangerer grensekostnaden de totale- og variable gj.snittkostnadene i bunnpunktet.

```python
# Definerer funksjonen.
def x(N,K):
    return (N**.5)*(K**.5)
x(N,K)
```

```python
# Lagrange likning.
L=w*N + r*K-lmda*(x(N,K)-X)
```

```python
# Deriver mhp N og K
dL_N=sp.diff(L,N)
dL_K=sp.diff(L,K)
foc_N=sp.Eq(dL_N,0)
foc_K=sp.Eq(dL_K,0)
display (foc_N)
display (foc_K)
```

```python
# Skriver den rel. prisen på arbeidskraft (w/r) som:
rel_w=(dL_N.args[0]/dL_K.args[0])

# og mtsb som:
mtsb=(dL_N.args[1]/dL_K.args[1])

# Sett disse lik som i likning (7).
optimal=sp.Eq(rel_w,mtsb)
display(optimal)
```
