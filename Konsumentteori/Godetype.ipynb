#### Normale og mindreverdige goder.

**Normale goder** er goder vi kjøper mer av når *inntekten* øker og mindre av når *inntekten* reduseres.
<br> **Mindreverdige goder** er goder vi kjøper mindre av når *inntekten* øker og mer av når *inntekten* reduseres.
<br> Godetypen identifiseres etter hvordan etterspørselen etter godet endrer seg når *inntekten* endres.r

```python
# Definerer funksjonen for normal gode.
def no(x):
    return 3.0e+6/(120+x)

# Definerer funksjonen for mindreverdig gode.
def mi(x):
    return 3.0e+6/(120+x)
```

```python
# Definerer x mellom tallene 0-500, med 500 punkter i intervallet.
x = np.linspace(0,500,500)
```

```python
# Definerer at jeg ønsker to figurer ved siden av hverandre.
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(20,5))

# Gir aksene navn.
ax1.set_ylabel('Pris')
ax1.set_xlabel('Etterspørsel')

ax2.set_ylabel('Pris')
ax2.set_xlabel('Etterspørsel')

# Tegner funksjonenen med forskjellige verdier.
ax1.plot(x,no(x),      color = "blue",  label= "Normale goder")
ax1.plot(x,no(x + 50), color = "red",   label= "Inntekten reduseres", linestyle = "--")
ax1.plot(x,no(x - 50), color = "green", label= "Innetekten øker",     linestyle = "--")

ax2.plot(x,mi(x),      color = "blue",  label= "Mindreverdige goder")
ax2.plot(x,mi(x + 50), color = "green", label= "Inntekten øker",      linestyle = "--")
ax2.plot(x,mi(x - 50), color = "red",   label= "Inntekten reduseres", linestyle = "--")

# Begrense aksene.
ax1.set(xlim=(0,300),ylim=(0,25000))
ax2.set(xlim=(0,300),ylim=(0,25000))

# Tar bort linjene øverst og til høyre.
ax1.spines['top'].set_color('none')
ax1.spines['right'].set_color('none')

ax2.spines['top'].set_color('none')
ax2.spines['right'].set_color('none')

# Gir figuren en tittel.
fig.suptitle('Inntekten forandres')

# Gir figuren en legend.
ax1.legend(loc='upper right')
ax2.legend(loc='upper right')

# Viser figuren.
plt.show()
```

På figurene over kan vi se at når *inntektene* øker så blir etterspøselen for **normale goder** økt mens etterspøselen for **mindreverdige goder** reduseres.
<br> Vi kan også se at det motsatte skjer når *inntektene* reduseres, da blir etterspørselen for **normale goder** reduseres mens etterspørselen for **mindreverdige goder** øker.

#### Komplementære goder og subtitutter.

**Komplementære goder**: Komplementer, også kalt komplentære produkter, er to eller flere produkter som samme oppfyller et bestemt behov, men som ikke nødvendigvis oppfyller behovet alene. 
<br> Eksempler på komplementære goder: pizza og dressing, bil og drivstoff, ski og staver.
<br> **Subtitutter**: Et substitutt er en erstaning, eller et alternativ for ett produkt eller en tjeneste.
<br> Eksempler på subtitutter: calzone og pizza, poteter og ris.
<br> **Godetypen** identifiseres etter hvordan etterspørselen for et gode endres når prisen på et annet gode endres.

```python
# Definerer funksjonen for komplementære goder.
def kg(x):
    return 3.0e+6/(120+x)

# Definerer funksjonen for subtitutter.
def st(x):
    return 5000 + (x**2)
```

```python
# Definerer x mellom tallene 0-500, med 500 punkter i intervallet.
x = np.linspace(0,500,500)
```

```python
# Definerer at jeg ønsker to figurer ved siden av hverandre.
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(20,5))

# Gir aksene navn.
ax1.set_ylabel('Prisen på pizza')
ax1.set_xlabel('Etterspørsel på dressing')

ax2.set_ylabel('Prisen på pizza')
ax2.set_xlabel('Etterspørsel på calzone')

# Tegner funksjonenen med forskjellige verdier.
ax1.plot(x,kg(x),        color = "blue",   label = "Etterspørsel på dressing")
ax1.vlines(50, 0, 17700, color = "red",    label = "Høy pris på pizza, lav etterspørsel")
ax1.vlines(250, 0, 8200, color = "green",  label = "Lav pris på pizza, høy etterspørsel")

ax2.plot(x,st(x),         color = "orange", label= "Etterspørsel på calzone")
ax2.vlines(50, 0, 7500,   color = "red",    label = "Lav pris på pizza, lav etterspørsel")
ax2.vlines(125, 0, 20700, color = "green",  label = "Høy pris på pizza, høy etterspørsel")

# Begrense aksene.
ax1.set(xlim=(0,300),ylim=(0,25000))
ax2.set(xlim=(0,300),ylim=(0,25000))

# Tar bort linjene øverst og til høyre.
ax1.spines['top'].set_color('none')
ax1.spines['right'].set_color('none')

ax2.spines['top'].set_color('none')
ax2.spines['right'].set_color('none')

# Gir figuren en tittel.
fig.suptitle('Prisen på pizza forandres')

# Gir figuren en legend og grid.
ax1.legend(loc='upper right')
ax2.legend(loc='upper right')
ax1.grid()
ax2.grid()

# Viser figuren.
plt.show()
```

På figuren over kan vi se at når prisen på pizza er høy blir etterspørselen for det **komplimentære gode** lav, mens etterspørselen for det **subtituerende gode** blir høy.
<br> Vi kan se det motsatte når prisen på pizza er lav, da blir etterspørselen for det **komplimentære gode** høy,  mens etterspørselen for det **subtituerende gode** blir lav.
