#### Etterspørselsfunksjoner

**Etterspørserlsfunksjoner:** Beskriver hvor mye en konsument ønsker å kjøpe av et bestemt gode.
<br> Etterspørselsfunksjoner utledes fra konsumentens nyttefunksjon. Konsumentens preferanser vil påvirke hvordan etterspørselsfunksjonen ser ut.
<br> $U(x_1,x_2) \rightarrow etterspørselsfunksjonen \leftarrow$ Vi foretrekker alle forskjellig, slik at vi har ulike etterspørselsfunksjoner. 
<br> Felles for alle etterspørselsfunksjoner:
<br> Kombinasjon av $p_1, p_2$ og $m$.
<br>
<br> **Eksempel (1)** på etterspørselsfunksjon for vare $x_1$:
<br> Vi sier at prisen for vare $x_1 = 10$ og inntekten $m = 100$.
<br> $x_1 = \frac{m}{p_1} = \frac{100}{10} = 10$
<br> Konsumenten etterspør altså 10 enheter av vare $x_1$.
<br>
<br> **Eksempel (2)** på etterspørselsfunksjon for vare $x_1$:
<br> Vi sier at prisen for vare $x_1 = 10$ og vare $x_2 = 50$.
<br> $x_1 = \frac{p_2}{p_1} = \frac{50}{10} = 5 $
<br> Konsumenten etterspør altså 5 enheter av vare $x_1$.
<br>
<br> Vi skal nå se på en etterspørselsfunksjon:

```python
def demand(p_1, p_2):
    return p_2 / p_1
```

```python
p_1 = np.linspace(1,10,100)
x_1 = np.linspace(0,5.55,7)
x_2 = np.linspace(0,8.3,7)
y_1 = np.linspace(9, 9, 7)
y_2 = np.linspace(6, 6, 7)
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Prisen på vare x_1")
ax.set_xlabel("Antall vare x_1")

# Plotter grafer, linjer og tekst.
ax.plot(p_1, demand(p_1, 50), color = "orange", label = "Etterspørsel")
ax.plot(x_1, y_1,             color = "red",    label = "Høy pris / Lav etterspørsel", linestyle = "--")
ax.plot(x_2, y_2,             color = "green",  label = "Lav pris / Høy etterspørsel", linestyle = "--")
ax.plot(5.55, 9,              color = "red",                                           marker = "o")
ax.plot(8.30, 6,              color = "green",                                         marker = "o")
ax.vlines(5.55, 0, 9,         color = "red",                                           linestyle = "--")
ax.vlines(8.30, 0, 6,         color = "green",                                         linestyle = "--")
ax.text(7.2,7.2, "MBV", color = "orange")

# Begrenser aksene. 
ax.set(xlim = (4,10))
ax.set(ylim = (4,10))

# Gir figuren en tittel.
ax.set_title("Etterspørselsfunksjon for vare x_1")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```

<br> Vi leser grafen fra venstre: "Desto lavere prisen på vare x_1 er, desto mer av vare x_1 kjøper vi".
<br> Etterspørselskurven representerer Marginal betalingsvilje (MBV): *Betalingsvillighet for en enhet til*.
<br> Det vil si at når prisen er veldig høy, må prisene falle ganske markant for at konsumenten skal være villig til å kjøpe en enhet ekstra.
<br> Det vil også si at når prisen er veldig lav, må ikke prisene falle like mye for at konsumenten skal være villig til å kjøpe en enhet ekstra.

**Oppgave**
<br> Anta at nyttefunksjonen til to ulike konsumemter er representer ved:
<br> $U_a(x_1,x_2) = x_1x_2$
<br> $U_b(x_1,x_2) = x_1^{0.5} + x_2$
<br> Finn et utrykk for disse konsumentenes etterspørsel etter vare $x_1$ og $x_2$.
<br>
<br> For å finne ut av dette må vi se på formlene for nyttemaksimering:
<br> 1) Budsjettbetingelsen: $p_1x_1 + p_2x_2 = m$
<br> 2) Tangeringsbetingelsen: $MSB = \frac{p_1}{p_2} = \frac{U'x_1}{U'x_2}$
<br>
<br> Vi starter med konsument ($U_a$):
<br> Beregne MSB: 
<br> $MSB = \frac{U'x_1}{U'x_2} = \frac{1}{1} * \frac{x_2}{x_1} = \frac{x_2}{x_1}$
<br> Setter MSB inn i tangeringsbetingelsen: 
<br> $MSB = \frac{p_1}{p_2}$
<br> $\frac{x_2}{x_1} = \frac{p_1}{p_2}$
<br> For å få $x_2$ alene på den ene siden av likhetstegner ganger vi med $x_1$: 
<br> $\frac{x_2}{x_1} = \frac{p_1}{p_2} |*x_1$
<br> $x_2 = \frac{p_1}{p_2}*x_1$
<br> Vi setter nå utrykket for $x_2$ inn i budsjettbetingelsen:
<br> $p_1x_1 + p_2x_2 = m$
<br> $p_1x_1 + p_2(\frac{p_1}{p_2}*x_1) = m$
<br> $p_1x_1 + p_1x_1 = m$
<br> $2p_1x_1 = m$
<br> Forenkler utrykket med å dele på $2p_1$, på begger sider av likhetstegnet:
<br> $2p_1x_1 = m |:2p_1$
<br> $x_1 = \frac{m}{2p_1}$
<br> <ins>Dette er altså utrykket for etterspørselsfunksjonen for vare $x_1$ til denne konsumenten.</ins>
<br> Vi setter nå inn utrykket for $x_1$ inn i utrykket for $x_2$ som vi fant tidligere:
<br> Utrykket vi fant isted: $x_2 = \frac{p_1}{p_2}*x_1$
<br> $x_2 = \frac{p_1}{p_2}*\frac{m}{2p_1}$
<br> $x_2 = \frac{m}{2p_2}$
<br> <ins>Dette er altså utrykket for etterspørselsfunksjonen for vare $x_2$ til denne konsumenten.</ins>
<br>
<br> Vi skal nå se på den andre konsumenten ($U_b$):
<br> Beregne MSB: 
<br> $MSB = \frac{U'x_1}{U'x_2}$
<br> $U'x_1 = 0.5x_1^{0.5 - 1} = 0.5x_1^{-0.5} = \frac{0.5}{x_1^{0.5}} = \frac{1}{2x_1^{0.5}}$
<br> $U'x_2 = 1x_2^{1-1} = 1x_2^{0} = 1 * 1 = 1$
<br> $MSB = \frac{\frac{1}{2x_1^{0.5}}}{1} = \frac{1}{2x_1^{0.5}}$
<br> Setter MSB inn i tangeringsbetingelsen: 
<br> $MSB = \frac{p_1}{p_2}$
<br> $\frac{1}{2x_1^{0.5}} = \frac{p_1}{p_2}$
<br> $\frac{2x_1^{0.5}}{1} = \frac{p_2}{p_1}$
<br> $2x_1^{0.5} = \frac{p_2}{p_1}$
<br> Forekler utrykket med å dele på 2 eller gange med $\frac{1}{2}$ på begge sider av likhetstegnet:
<br> $2x_1^{0.5} = \frac{p_2}{p_1} |:2$
<br> $x_1^{0.5} = \frac{p_2}{2p_1}$
<br> For å bli kvitt eksponenten med 0.5 (kvadratroten) opphøyer vi utrykkene i 2:
<br> $(x_1^{0.5})^{2} = (\frac{p_2}{2p_1})^{2}$
<br> $x_1 = \frac{p_2^2}{(2p_1)^2}$
<br> $x_1 = \frac{p_2^2}{2^2p_1^2}$
<br> $x_1 = \frac{p_2^2}{4p_1^2}$
<br> <ins>Dette er altså utrykket for etterspørselsfunksjonen for vare $x_1$ til denne konsumenten.</ins>
<br> Siden utrykket for etterspørselen for gode $x_2$ forsvant når vi partiellderiverte må vi bruke budsjettbetingelsen for å finne funksjonen for gode $x_2$.
<br> Vi setter inn utrykket for gode $x_1$ inn i budsjettbetingelsen:
<br> $p_1x_1 + p_2x_2 = m$
<br> $p_1* \frac{p_2^2}{4p_1^2} + p_2x_2 = m$
<br> $\frac{p_2^2}{4p_1} + p_2x_2 = m$
<br> Siden målet er å få $x_2$ alene på den ene siden av likhetstegnet kan vi flytte hele det første leddet på andre siden og bytte fortegn.
<br> $p_2x_2 = m - \frac{p_2^2}{4p_1}$
<br> Deler på $p_2$ for å få $x_2$ alene.
<br> $p_2x_2 = m - \frac{p_2^2}{4p_1} | :p_2$
<br> $x_2 = \frac{m}{p_2} - \frac{p_2^2}{4p_1p_2}$
<br> $x_2 = \frac{m}{p_2} - \frac{p_2}{4p_1}$
<br> <ins>Dette er altså utrykket for etterspørselsfunksjonen for vare $x_2$ til denne konsumenten.</ins>

#### Etterspørsel og godetype

**Normale goder**
<br> Goder du kan kjøpe mer av når inntekten øker og mindre av når inntekten reduseres.
<br> Se for deg at du kan velge mellom brus som er gode $x_1$ og nuddler som er gode $x_2$.
<br> Vi skal nå se på etterspørselsfunksjonen til brus og avgjøre om det er ett normalt gode.
<br> Vi antar at etterspørselfunksjonen til gode $x_1$ er gitt ved: 
<br> $x_1 = \frac{m}{p_1}$
<br> I dette eksempelet bruker vi verdiene:
<br> $m = 100$ og $p_1 = 20$.
<br> Da vil etterspurt kvantum være like:
<br> $x_1 = \frac{m}{p_1} = \frac{100}{20} = 5$
<br> Vi skal nå se på hva som skjer når inntekten øker for å avgjøre om det er ett normalt gode.
<br> I dette eksempelet bruker vi verdiene:
<br> $m = 200$ og $p_1 = 20$.
<br> Da vil etterspurt kvantum være like:
<br> $x_1 = \frac{m}{p_1} = \frac{200}{20} = 10$
<br> Som vi kan se når inntekten øker, øker også etterspørselen. Derfor kan vi si at gode $x_1$ altså brus er et normalt gode.
<br> Vi skal nå tegne opp etterspørselsfunksjonen til gode $x_1$ for å se nærmere på det.

```python
# Definerer etterspørselsfunksjonen.
def dem(m, p):
    return m / p
```

```python
x_ax = np.linspace(1,400,1000)
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Pris vare x_1")
ax.set_xlabel("Mengde vare x_1")

# Plotter budsjettlinjen
ax.plot(x_ax, dem(200,x_ax), color = "blue",   label = "Etterspørsel gode x_1")
ax.plot(x_ax, dem(300,x_ax), color = "green",  label = "Økt inntekt") 
ax.plot(x_ax, dem(100,x_ax), color = "red",    label = "Redusert inntekt") 

# Begrenser aksene. 
ax.set(xlim = (0,30))
ax.set(ylim = (0,30))

# Gir figuren en tittel.
ax.set_title("Etterspørsel")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```

Fra figuren over kan vi se at når inntekten øker gjør etterspørselsfunksjonen ett positivt skift, og motsatt når inntektene reuseres.
<br> Vi kan se at når prisen på vare $x_1$ er f.eks 20, etterpurte konsumenten orginalt 10 enheter, når inntekten øker etterspør konsumenten nå 15 enheter, og motsatt går etterspørselen ned til 5 enheter når inntekten reduseres.

**Mindreverdige goder**
<br> Goder du kan kjøpe mindre av når inntekten øker og mer av når inntekten reduseres.
<br> Da skal vi se vidre på nuddlene i forrige eksempel, altså gode $x_2$
<br> Vi skal nå se på etterspørselsfunksjonen til nuddler og avgjøre om det er ett mindreverdig gode.
<br> Vi antar at etterspørselfunksjonen til gode $x_2$ er gitt ved: 
<br> $x_2 = \frac{1000}{m*p_2}$
<br> I dette eksempelet bruker vi verdiene:
<br> $m = 100$ og $p_2 = 5$.
<br> Da vil etterspurt kvantum være like:
<br> $x_2 = \frac{1000}{m*p_2} = \frac{1000}{100*5} = \frac{1000}{500} = 2$
<br> Vi skal nå se på hva som skjer når inntekten øker for å avgjøre om det er ett mindreverdig gode.
<br> I dette eksempelet bruker vi verdiene:
<br> $m = 200$ og $p_2 = 5$.
<br> Da vil etterspurt kvantum være like:
<br> $x_2 = \frac{1000}{m*p_2} = \frac{1000}{200*5} = \frac{1000}{1000} = 1$
<br> Som vi kan se når inntekten øker, synker etterspørselen. Derfor kan vi si at gode $x_2$ altså nuddler er et mindrevrdig gode.
<br> Vi skal nå tegne opp etterspørselsfunksjonen til gode $x_2$ for å se nærmere på det.

```python
# Definerer etterspørselsfunksjonen.
def dem_nud(m, p):
    return 1000 / (m * p)
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Pris vare x_2")
ax.set_xlabel("Mengde vare x_2")

# Plotter budsjettlinjen
ax.plot(x_ax, dem_nud(200,x_ax), color = "blue",   label = "Etterspørsel gode x_1")
ax.plot(x_ax, dem_nud(300,x_ax), color = "green",  label = "Økt inntekt") 
ax.plot(x_ax, dem_nud(100,x_ax), color = "red",    label = "Redusert inntekt") 

# Begrenser aksene. 
ax.set(xlim = (1,5))
ax.set(ylim = (0,5))

# Gir figuren en tittel.
ax.set_title("Etterspørsel")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```

Fra figuren over kan vi se at når inntekten øker gjør etterspørselsfunksjonen ett negativt skift, og motsatt når inntektene reuseres.
<br> Vi kan se at når prisen på vare $x_2$ er f.eks 2, etterpurte konsumenten orginalt 2.5 enheter, når inntekten øker etterspør konsumenten nå 1.7 enheter, og motsatt går etterspørselen til 5 enheter når inntekten reduseres.
<br> Merk at når vi forandrer en variabel som ikke er representert på grafen, vil vi får en nye graf. 

**Komplementære goder** 

Goder du kjøper mindre av når prisen på det komplimentære gode øker og kjøper mer av når prisen på det komplimentære gode reduseres.
<br> F.eks. gode rømmedressing ($x_1$) og gode pizza ($x_2$) er to komplimentære goder. Vi skal senere også se på gode calzone ($x_3$).
<br> Vi tenker oss at etterspørselsfunksjonen til gode $x_1$ er gitt ved:
<br> $x_1 = \frac{100}{p_2}$
<br> Gode $x_1$ er altså avhengig av prisen på gode $x_2$. La oss si at $p_2 = 50$
<br> Da vil etterspørselene til gode $x_1$ se slik ut:
<br> $x_1 = \frac{100}{50} = 2$
<br> Men så øker prisen på gode $x_2$ til $p_2 = 100$
<br> Da vil etterspørselene til gode $x_1$ se slik ut:
<br> $x_1 = \frac{100}{100} = 1$
<br> Vi kan se at rømmedressing er ett komplimentært gode til pizza.
<br> Vi skal nå se videre på dette i ett pris/mengde diagram.

```python
# Definerer etterspørselsfunksjonen.
def dem_dres(p):
    return 100 / p
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Pris vare x_2")
ax.set_xlabel("Mengde vare x_1")

# Plotter budsjettlinjen
ax.plot(x_ax, dem_dres(x_ax), color = "blue",  label = "Etterspørsel gode x_1")
ax.plot(10, 10,               color = "green", label = "Orginal"              , marker = "o")
ax.plot(5,  20,               color = "red",   label = "Ny"                   , marker = "o")

# Begrenser aksene. 
ax.set(xlim = (0,20))
ax.set(ylim = (0,20))

# Gir figuren en tittel.
ax.set_title("Etterspørsel")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```

Vi kan se her at i motsettning til de forrige etterspørselsfunksjonen vi har sett på har vi mengde vare $x_1$ på x-aksen men prisen på vare $x_2$ på y-aksen.
<br> Vi kan se at orginalt var prisen på pizza 10 kr, da etterspurte vi 10 enheter rømmedressing. Men så blir prisen på pizza skrudd opp til 20 kr, og da etterspør vi bare 5 ehnheter rømmedressing. 
<br> Merk at når vi forandrer en variabel som er representert på grafen, vil vi skli langs den eksisterende grafen. 

**Substitutt**
<br> Gode du kjøper mer av når prisen på substituttet øker og kjøper mindre av når prisen på substituttet synker.
<br> Som nevnt tidligere skulle vi også se på calzone, altså gode $x_3$.
<br> Vi tenker oss at etterspørselsfunksjonen til gode $x_3$ er gitt ved:
<br> $x_3 = \frac{p_2}{p_3}$
<br> Gode $x_3$ er altså avhengig av prisen på gode $x_2$. La oss si at $p_2 = 50$ og $p_3 = 50$.
<br> Da vil etterspørselene til gode $x_3$ se slik ut:
<br> $x_3 = \frac{50}{50} = 1$
<br> Men så øker prisen på gode $x_2$ til $p_2 = 100$
<br> Da vil etterspørselene til gode $x_3$ se slik ut:
<br> $x_3 = \frac{100}{50} = 2$
<br> Vi kan se at calzone er ett substitutt til gode pizza.
<br> Vi skal nå se videre på dette i ett diagram.

```python
# Definerer etterspørselsfunksjonen.
def dem_cal(p_2, p_3):
    return p_2 / p_3
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Pris vare x_2")
ax.set_xlabel("Mengde vare x_3")

# Plotter budsjettlinjen
ax.plot(x_ax, dem_cal(x_ax, 50), color = "blue",  label = "Etterspørsel gode x_3")
ax.plot(100, 2,               color = "green", label = "Orginal"              , marker = "o")
ax.plot(200, 4,               color = "red",   label = "Ny"                   , marker = "o")

# Begrenser aksene. 
ax.set(xlim = (0,400))
ax.set(ylim = (0,10))

# Gir figuren en tittel.
ax.set_title("Etterspørsel")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```

Vi kan se her at desto høyere prisen på vare $x_2$ er, øker etterspørselen på vare $x_3$.
<br> Orginalt vare prisen på pizza 2 kr og da etterspurte konsumenten 100 calzone, men når prisen på pizza gikk opp til 4 kr, økte etterspørselen til konsumenten til 200 calzone.
