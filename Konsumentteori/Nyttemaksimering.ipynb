**Tangeringsbetingelsen**
<br> *Indifferenskurven tangerer med budsjettskurven.*
<br> Konsumenten vil tilpasse seg til det punktet vedkommende har st√∏rst nytte gitt sitt budsjett. 

```python
# Definerer en funksjon for budsjettkurven
def bud_1(x):
    return 20 - 0.5 * x

# Definerer indifferenskurve 1
def ind_1(x):
    return (10**(5/3))/x**(2/3)

# Definerer indifferenskurve 2
def ind_2(x):
    return (13.75**(5/3))/x**(2/3)
```

```python
# Definerer to intervaller mellom 0/1 - 100, med 100 punkter.
x_1 = np.linspace(0, 100, 100)
x_2 = np.linspace(1, 100, 100)
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Vare x_2")
ax.set_xlabel("Vare x_1")

# Plotter grafer, linjer og tekst.
ax.plot(x_1, bud_1(x_1), color = "blue",   label = "Budsjettkurve")
ax.plot(x_1, ind_1(x_2), color = "green",  label = "Indiffirenskurve (n = 5)")
ax.plot(x_1, ind_2(x_2), color = "orange", label = "Indiffirenskurve (n = 13.75)")
ax.plot(9.65, 9.65,      color = "black", marker = "o")
ax.plot(30.5, 4.7,       color = "black", marker = "o")
ax.plot(15, 12.5,        color = "black", marker = "o")
ax.text(10,10,  "A")
ax.text(31,5.1, "B")
ax.text(15,13,  "C")


# Begrenser aksene. 
ax.set(xlim = (0,50))
ax.set(ylim = (0,25))

# Gir figuren en tittel.
ax.set_title("Tangeringsbetingelse")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```

P√• figuren over kan vi se:
<br> <span style='color:Blue'> Budsjettkurve: $x_2 = \frac{m}{p_2} - \frac{p_1x_1}{p_2}$
<br> Omr√•det nedenfor budsjettkurven kalles mulighetsomr√•det og konsumenten har mulighet til alle godekombinasjoner i dette omr√•det. Omr√•det ovenfor budsjettkurven har konsumenten ikke r√•d til. Selve budsjettkurven viser hva konsumenten maksimalt har r√•d til.
<br> <span style='color:Green'> P√• indifferenskurven kan vi se at konsumenten har valgt √• tilpasset seg i godekombinasjonen punkt A. Men er dette den optimale godekombinasjonen? Konsumenten har r√•d til denne godekombinasjonen, men kan kj√∏pe mer av b√•de vare $x_1$ og vare $x_2$. Visst konsumenten tar denne bemerkningen til etterettning og velger √• tilpasse seg i godekombinasjonen punkt B f√•r vedkommende brukt opp hele budsjettet. Men siden indifferenskurven viser den n√∏yaktig samme nytten hele veien har ikke dette noen hensikt. Alts√• opplever konsumenten den samme nytten i punkt A som i punkt B bare at vedkommende f√•r brukt opp hele budsjettet i punkt B. 
<br> <span style='color:Orange'> Marginalsubstitusjonsbr√∏k: $MSB = \frac{p_1}{p_2}$
<br> Konsumenten √∏nsker √• oppn√• s√• h√∏y nytte som mulig gitt budsjettet. Denne konsumenten kan n√• ett h√∏yere nytteniv√• ved √• velge kodekobinasjonen hvor indifferenskurven tangerer med budsjettkurven. I tangeringspunktet medg√•r hele inntekten til godekombinasjonen og samtidig opplever konsumenten den h√∏yeste graden av mulig nytte.
