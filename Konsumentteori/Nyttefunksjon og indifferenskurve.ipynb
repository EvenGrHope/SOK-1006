## Nyttefunksjon og indifferenskurve

```python
# Laster inn pakker.
import numpy as np
from sympy import *
from matplotlib import pyplot as plt
```

<hr/>

**Nyttefunksjon:**
<br> Konsumenten foretrekker noen godekombinasjoner framfor andre, dette avhenger av prisen, inntekten og *konsumentens preferanser*.
<br> **Nyttefunksjonen** representerer *konsumentens preferanser*. $U(x_1,x_2) = $ *Konsumentens preferanser*.
<br> Med **nyttefunksjonen** kan vi beregne en bestemt verdi av opplevd behovstilfredsstillelse for en gitt mengde konsum av vare $x_1$ og vare $x_2$. Det er en m책leenhet p책 hvor bra konsumenten har det.
<br> $U = 20 > U = 15$
<br> **Nyttefunksjonen** er individuell for alle konsumenter.
<br>
<br> Vi skal n책 se p책 en nyttefunksjon der vare $x_2$ er konstant. 

```python
# Definerer nyttefunksjonen.
def ny(x):
    return x**0.5
```

```python
# Definerer x mellom tallene 0-10, med 100 punkter i intervallet.
x   = np.linspace(0,100,500)
x_1 = np.linspace(0,6.2,10)
x_2 = np.linspace(2.5,2.5,10)
x_3 = np.linspace(0,1,10)
x_4 = np.linspace(1,1,10)
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Nytte")
ax.set_xlabel("Vare x_1")

# Plotter budsjettlinjen
ax.plot(x, ny(x), color = "blue",   label = "Nyttefunksjon")
ax.plot(x_1, x_2, color = "orange", label = "Lav marginalnyte") 
ax.plot(x_3, x_4, color = "purple", label = "H첩y marginalnyte") 

# Begrenser aksene. 
ax.set(xlim = (0,10))
ax.set(ylim = (0,3))

# Gir figuren en tittel.
ax.set_title("Nyttefunksjon")

# Gir figuren en legend.
ax.legend(loc = "lower right")

# Viser figuren.
plt.grid()
plt.show()
```

<hr/>

N책 skal vi se p책 **nyttefunksjoner** der vare $x_2$ ikke er konstant. Vi har forskjellige typer **nyttefunksjoner**, Cobb-Douglas og Kvasiline챈r.
<br> **Cobb-Douglas:**
<br> $U(x_1,x_2) = x_1x_2$
<br> $U(x_1,x_2) = x_1^{0.5}x_2^{0.5}$
<br> **Kvasiline챈r:**
<br> $U(x_1,x_2) = x_1 + 2x_2^{0.5}$

<hr/>

**Eksempel med Cobb-Douglas:**
<br> $U(5,2) = 5 * 2$
<br> $U(5,2) = 10$
<br>
<br> $U(10,1) = 10 * 1$
<br> $U(10,1) = 10$

Vi kan se at konsumenten i dette tilfelle er **indifferent** om hvilket godekombinasjoner individet m책 ha, s책 lenge nytten er den samme.

<hr/>

**Indifferens kurve**

Vi skal n책 tegne opp indifferenskurver.

```python
# Definerer nyttefunksjonen.
def ind_1(x_0):
    return (4000**(5/3))/x_0**(2/3)

def ind_2(x_0):
    return (4741**(5/3))/x_0**(2/3)

def ind_3(x_0):
    return (5500**(5/3))/x_0**(2/3)
```

```python
x_0 = np.linspace(1,10500,100000)
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Vare x_2")
ax.set_xlabel("Vare x_1")

# Plotter budsjettlinjen.
ax.plot(x_0, ind_2(x_0), color = "blue",  label = "Orginal")
ax.plot(x_0, ind_1(x_0), color = "red",   label = "Lavere nytte", linestyle = "--")
ax.plot(x_0, ind_3(x_0), color = "green", label = "H첩yere nytte", linestyle = "--")

# Begrenser aksene. 
ax.set(xlim = (0,10000))
ax.set(ylim = (0,10000))

# Gir figuren en tittel.
ax.set_title("Indifferenskurve")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```

**Indifferenskurvene** viser oss at konsumenten er indifferent mellom godekombinasjonenen p책 samme graf, siden de har samme nytte i forskjellige kombinasjoner.
