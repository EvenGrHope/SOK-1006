## Budsjettkurve

Budsjettkurven viser de forbrukskombinasjonene som koster akkurat like mye som konsumentens inntekt.

```python
# Laster inn pakker
import numpy as np
from sympy import *
from matplotlib import pyplot as plt
```

<hr/>

#### Utlede budsjettkurven

Et individ har inntekten gitt ved $m$. Individetet står ovenfor prisen gitt ved $p_1$ og $p_2$ på henholdvis vare $x_1$ og $x_2$.
<br> $m$ = Inntekt/Budsjett
<br> $p_1$ = Prisen på vare 1
<br> $p_2$ = Prisen på vare 2
<br> $x_1$ = Antall vare 1
<br> $x_2$ = Antall vare 2
<br> I dette eksempelet er $p_1 = 5$, $p_2 = 10$ og $m = 100$
<br> Budsjettkurven er gitt ved $p_1x_1 + p_2x_2 = m$

For å kunne lage en funksjon for budsjettlinjen må vi finne ut hvor mange av vare $x_2$ konsumenten kan kjøpe.
<br> Det finner vi ut slik:
<br> $p_1x_1 + p_2x_2 = m $
<br>
<br> $p_2x_2 = m - p_1x_1 $
<br>
<br> $\frac{p_2x_2}{p_2} = \frac{m}{p_2} - \frac{p_1x_1}{p_2} $
<br>
<br> $x_2 = \frac{m}{p_2} - \frac{p_1x_1}{p_2} $
<br>
<br> En rett linje skrives som $y = b+ax$. Det er altså det samme som vi har over:
<br> $y = b+ax$
<br> $x_2 = \frac{m}{p_2} - \frac{p_1x_1}{p_2} $

Med prisene i eksempelet over kan vi finne ut hvor mange varer konsumenten kan kjøpe av vare $x_2$:
<br> $x_2 = \frac{100}{10} - \frac{5x_1}{10}$
<br>
<br> $x_2 = 10 - 0.5x_1$
<br> Konsumenten må altså gi fra seg $0.5$ av vare $x_2$ for å få en ekstra av vare $x_1$.
<br> Vi kan også se at visst konsumenten bruker hele sitt busjett på vare $x_2$ vil ha få $10$ enheter.

La oss nå tegne opp budsjettlinjen.

```python
# Definerer prisen på varene og budsjettet.
p_1 = 5
p_2 = 10
m   = 100

# Definerer x_1 mellom tallene 0-20, med 100 punkter i intervallet.
x_1 = np.linspace(0,20,100)
```

```python
# Lager funksjonen for budsjettlinjen.
def bud_1(x_1):
    return 10 - 0.5 * x_1
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Vare x_2")
ax.set_xlabel("Vare x_1")

# Plotter budsjettlinjen
ax.plot(x_1, bud_1(x_1), color = "blue",    label = "Budsjettlinje")

# Begrenser aksene. 
ax.set(xlim = (0,25))
ax.set(ylim = (0,15))

# Gir figuren en tittel.
ax.set_title("Budsjettlinje")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```

Den blå linjen kaller vi *budsjettlinjen*.
<br> Området under budsjettlinjen kaller vi *konsumentens mulighetsområdet*.
<br> Området over budsjettlinjen har ikke konsumenten mulighet til å oppnå med mindre budsjette øker eller prisene faller.
<br> Konsumenten kan altså kjøpe alle varekombinasjoner på og under budsjettlinjen, med sitt gitte budsjett.

#### Markedsmessig bytteforhold

Det markedsmessige bytteforholdet mellom vare 1 og vare 2 forteller oss hvor mange enheter av den ene varen vi som opptrer i markedet går glipp av hvis vi kjøper en enhet til av den andre varen.

Forklar hva som menes med det markedsmessige bytteforholdet mellom vare $x_1$ og $x_2$.
<br> Benytt prisene $p_1 = 15$, $p_2 = 5$ og $m = 60$ til å utlede et eksempel som kan illustreres i et diagram.

Vi kan umidelbart se at vare $x_1$ er tre ganger så dyr som vare $x_2$. 
<br> Det vil si at vi må legge fra oss tre enheter av vare $x_2$ for å få råd til en enheter av vare $x_1$.
<br> Dette er altså bytteforholdet mellom varene.
<br> Bytteforholde er gitt ved følgende formel:
<br> $\frac{p_1}{p_2}$
<br>
<br> $\frac{p_1}{p_2} = \frac{15}{5} = 3$

La oss tegne opp det markedsmessige bytteforholdet.

```python
# Definerer prisen på varene og budsjettet.
p_1 = 15
p_2 = 5
m   = 60

# Definerer x_2 mellom tallene 0-20, med 100 punkter i intervallet.
x_2 = np.linspace(0,20,100)
```

```python
# Lager funksjonen for budsjettlinjen.
def bud_2(x_1):
    return 12 - 3 * x_1
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Vare x_2")
ax.set_xlabel("Vare x_1")

# Plotter budsjettlinjen
ax.plot(x_2, bud_2(x_2), color = "blue",    label = "Bytteforhold")

# Begrenser aksene. 
ax.set(xlim = (0,5))
ax.set(ylim = (0,14))

# Gir figuren en tittel.
ax.set_title("Markedsmessig bytteforhold")

# Gir figuren en legend.
ax.legend(loc = "upper right")
ax.grid()

# Viser figuren.
plt.show()
```

Her kan vi altså se at når vi kjøper en ekstra vare $x_1$ må vi gi opp tre av vare $x_2$.

#### Anvende budsjettkurven

Vi skal nå anvend budsjettkurven til å løse noen oppgaver.
<br>Et individ har inntekten $m = 200$, og individet står ovenfor prisen $p_1 = 25$ og $p_2 = 10$.

Oppgaver:
<br> 1) Hvor mange enheter av vare $x_1$ har individet maksimalt anledning til å kjøpe?
<br> 2) Hvor mange enheter av vare $x_2$ har individet maksimalt anledning til å kjøpe?
<br> 3) Hva er bytteforholdet mellom vare $x_1$ og vare $x_2$?
<br> 4) Illustrer budsjettkurven til dette individet i et diagram.
<br> 5) Anvend budsjettkurven til å beregne hvor mange enheter av vare $x_2$ individet har anledning til å kjøpe, gitt at individet kjøper 6 av vare $x_1$. Tegn godekombinasjonen på budsjettkurven du illustrerte i oppgave 4.

Besvarelse:
<br> 1) Vi kan finne ut hvor mange varer av $x_1$ individet maksimalt kan kjøpe på følgende måte:
<br> $\frac{m}{p_1} = \frac{200}{25} = 8$
<br>
<br> 2) Vi kan finne ut hvor mange varer av $x_2$ individet maksimalt kan kjøpe på følgende måte:
<br> $\frac{m}{p_2} = \frac{200}{10} = 20$
<br>
<br> 3) Vi kan finne bytteforholde mellom varene på følgende måte:
<br> $\frac{p_1}{p_2} = \frac{25}{10} = 2.5$
<br> Vi må altså gi fra oss 2.5 enhet av vare $x_2$ for å får en vare ekstra av vare $x_1$.

```python
# Oppgave 4 og oppgave 5

# Definerer prisen på varene og budsjettet.
p_1 = 25
p_2 = 10
m   = 200

# Definerer x_3 mellom tallene 0-20, med 100 punkter i intervallet.
x_3 = np.linspace(0,20,100)
```

```python
# Lager funksjonen for budsjettlinjen.
def bud_3(x_1):
    return 20 - 2.5 * x_1
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Lager to dict. med kordinatene til figurativ framvisning. 
point_1 = [0, 5]
point_2 = [6, 5]

# Lager to nye dict. med kordinatene til figurativ framvisning.
x_values = [point_1[0], point_2[0]]
y_values = [point_1[1], point_2[1]]

# Gir aksene navn.
ax.set_ylabel("Vare x_2")
ax.set_xlabel("Vare x_1")

# Plotter budsjettlinjen og figurativ framstilling forutsatt at konsumenten kjøper 6 av vare x_1.
ax.plot(x_3, bud_3(x_3), color = "blue",    label = "Budsjettkurve")
ax.plot(x_values, y_values, color = "orange")
ax.vlines(6, 0, 5, color = "orange")

# Begrenser aksene. 
ax.set(xlim = (0,9))
ax.set(ylim = (0,22))

# Gir figuren en tittel.
ax.set_title("Budsjettkurve")

# Gir figuren en legend.
ax.legend(loc = "upper right")
ax.grid()

# Viser figuren.
plt.show()
```

Vi kan se at visst konsumenten kjøper 6 varer av $x_1$, kan konsumenten også kjøpe 5 varer av $x_2$.
<br> Vi kan også regne oss fram til hvor mange varer av $x_2$ konsumenten kan kjøpe gitt at konsumenten kjøper 6 av vare $x_1$:
<br>
<br> $p_1x_1 + p_2x_2 = m $
<br>
<br> $p_2x_2 = m - p_1x_1 $
<br>
<br> $\frac{p_2x_2}{p_2} = \frac{m}{p_2} - \frac{p_1x_1}{p_2} $
<br>
<br> $x_2 = \frac{m}{p_2} - \frac{p_1x_1}{p_2} $
<br>
<br> $x_2 = \frac{200}{10} - \frac{6*25}{10} $
<br>
<br> $x_2 = \frac{200}{10} - \frac{150}{10} $
<br>
<br> $x_2 = 20 - 15 $
<br>
<br> $x_2 = 5 $

#### Endring i budsjettkurve

Et individ har inntekt gitt ved $m$. Individet står ovenfor prisene gitt ved $p_1$ og $p_2$ på henholdsvis vare $x_1$ og $x_2$.
<br> Illustrer i ett diagram virkningen på budsjettkurven ved følgende endringer: 
<br> 1) $p_1$ og $p_2$ er uendret. Inntekten $m$ reduseres med 20%.
<br> 2) Proporsjonal reduksjon i begge prisene $p_1$ og $p_2$.
<br> 3) Inntekte $m$ er uendret. $p_1$ øker med 20%, $p_2$ reduseres med 20%.
<br> 4) $p_1$, $p_2$ og inntekten $m$ reduseres med 5%.
<br>
<br> Anta at $p_1 = 10, p_2 = 10, m = 100$

Besvarelse:
<br> 1) Vi starter med å regne ut hvor mye nytt budjett er:
<br> $m = 100 * 0.80 = 80$
<br>
<br> $x_2 = \frac{m}{p_2} - \frac{p_1x_1}{p_2} $
<br>
<br> $x_2 = \frac{80}{10} - \frac{10*x_1}{10} $
<br>
<br> $x_2 = 8 - x_1 $
<br>
<br> 2) Vi antar at prisene reduseres med 10%:
<br> $p_1 = 10 * 0.90 = 9$
<br> $p_2 = 10 * 0.90 = 9$
<br>
<br> $x_2 = \frac{m}{p_2} - \frac{p_1x_1}{p_2} $
<br>
<br> $x_2 = \frac{100}{9} - \frac{9*x_1}{9} $
<br>
<br> $x_2 = 11,11 - x_1 $
<br> 
<br> 3) Starter med å regne ut endring i prisene:
<br> $p_1 = 10 * 1.20 = 12$
<br> $p_2 = 10 * 0.80 = 8$
<br>
<br> $x_2 = \frac{m}{p_2} - \frac{p_1x_1}{p_2} $
<br>
<br> $x_2 = \frac{100}{8} - \frac{12*x_1}{8} $
<br>
<br> $x_2 = 12.5 - 1.5x_1 $
<br>
<br> 4) Starter med å regne ut endringene i prisene og budsjettet:
<br> $m = 100 * 0.95 = 95$
<br> $p_1 = 10 * 0.95 = 9.5$
<br> $p_2 = 10 * 0.95 = 9.5$
<br>
<br> $x_2 = \frac{m}{p_2} - \frac{p_1x_1}{p_2} $
<br>
<br> $x_2 = \frac{95}{9.5} - \frac{9.5*x_1}{9.5} $
<br>
<br> $x_2 = 10 - x_1 $

```python
# Definerer prisen på varene og budsjettet.
p_1 = 10
p_2 = 10
m   = 100

# Definerer x_3 mellom tallene 0-20, med 100 punkter i intervallet.
x_4 = np.linspace(0,20,100)
```

```python
# Lager funksjoner for budsjettlinjene.
def bud_4(x_1):
    return 10 - x_1

def bud_5(x_1):
    return 8 - x_1

def bud_6(x_1):
    return 11 - x_1

def bud_7(x_1):
    return 12.5 - 1.5*x_1
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Vare x_2")
ax.set_xlabel("Vare x_1")

# Plotter budsjettlinjene
ax.plot(x_4, bud_4(x_4), color = "blue",    label = "Orginal")
ax.plot(x_4, bud_5(x_4), color = "red",     label = "Oppgave 1", linestyle = "--")  
ax.plot(x_4, bud_6(x_4), color = "green",   label = "Oppgave 2", linestyle = "--") 
ax.plot(x_4, bud_7(x_4), color = "orange",  label = "Oppgave 3", linestyle = "--") 

# Begrenser aksene. 
ax.set(xlim = (0,12))
ax.set(ylim = (0,14))

# Gir figuren en tittel.
ax.set_title("Endring i budsjettkurve")

# Gir figuren en legend.
ax.legend(loc = "upper right")
ax.grid()

# Viser figuren.
plt.show()
```

Vi kan se at orignalen og oppgave 4 har samme kurve. Dette er fordi både prisene og budsjettet synker med det samme, altså 5%. Kjøpekraften er den samme.
