## Subtitusjon- og inntektseffekten

```python
# Laster inn pakker.
import numpy as np
from sympy import *
from matplotlib import pyplot as plt
```

**Subtitusjon- og inntektseffekten**
<br> Hvordan prisendring påvirker konsumentens etterspørsel etter godene $x_1$ og $x_2$.
<br> I dette eksempelet skal vi se på godene ris og potet.
<br> Konsumenten kan maksimalt kjøpe 10 av ris eller 10 av potet.

```python
# Definerer den orginale budsjettkurven.
def bud_01(x_1):
    return 10 - 1*x_1

# Definerer budsjettkurve med fordoblet rispris.
def bud_02(x_1):
    return 10 - 2*x_1
```

```python
# Definerer et intervall mellom 0 - 10, med 10 punkter.
x_1 = np.linspace(0,20,100)
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Potet")
ax.set_xlabel("Ris")

# Plotter grafer, linjer og tekst.
ax.plot(x_1, bud_01(x_1), color = "blue",   label = "Budsjettkurve 01 (B_1)")
ax.plot(x_1, bud_02(x_1), color = "red",    label = "Budsjettkurve 02 (B_2)")
ax.plot(7, 3,             color = "orange", label = "Optimal tilpasning 01 (A)", marker = "o")
ax.plot(2, 6,             color = "purple", label = "Optimal tilpasning 02 (B)", marker = "o")

# Begrenser aksene. 
ax.set(xlim = (0,12))
ax.set(ylim = (0,12))

# Gir figuren en tittel.
ax.set_title("Subtitusjon- og inntektseffekten")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```

Fra grafen over kan vi se effekten av hva som skjer når gode ris ble dobbelt så dyrt. Det har skjedd to ting:

**Substitusjonseffekten**
<br> Potet framstår som mer attraktivt enn tidligere fordi bytteforholdet mellom poteter og ris har drastitskt endret seg.
<br> Substitusjonseffekten er differansen mellom punkt A og punkt C (som ikke er tegnet inn) langs $x_1$-aksen.
<br> Punkt C finner vi med å paralell forskyve den nye budsjettkurven til den tangerer den oprinelige indifferenskurven.
<br> Effekten av at den relative prisen har endret seg. Nytten holdes konstant. 
<br> I dette eksempelet sier vi at konsumenten substituerer mot poteter. 
<br> Substitusjonseffekten kan forklares som hvordan denne konsumenten har valgt å endre sitt konsum av godekombinasjonen poteter og ris, dersom konsumenten ble kompansert med en inntektsøkning slik at konsumenten kunne vært på det samme nyttenivået som tidligere men med det nye prisnivået. 
<br> Substitusjonseffekten er altså en glidning langs den oprinelige indifferenskurven.

**Inntektseffekten**
<br> Kjøpekraften til konsumenten har blitt redusert. Konsumenten har ikke lengere råd til den samme godekombinasjonen som tidligere.'
<br> Inntektseffekten er differansen mellom punkt B og punkt C (som ikke er tegnet inn) langs $x_1$-aksen.
<br> Punkt C finner vi med å paralell forskyve den nye budsjettkurven til den tangerer den oprinelige indifferenskurven.
<br> Siden realinntekten er lavere har ikke konsumenten lengere råd til å ligge på den orginale idifferenskurven og konsumenten må tilpasse seg til ett lavere nyttenivå.
<br> Inntekteffekten kan forklare som hvordan konsumenten må endre sitt konsum av poteter og ris når inntekten reduseres slik at konsumenten lander på den nye budsjettkurven. Altså den nye indifferenskurven tangerer med den nye budsjettlinjen.

<br> Totaleffekten avhenger av hvorvidt inntektseffekten eller substitusjonseffekten dominerer for dette gode.

**Oppgave subtitusjons- og inntektseffekten**

Vi ser på et individ med inntekt $m = 48$. Vi lar prisene på de to varene være: 
<br> $p_1 = 6$ 
<br> $p_2 = 8$
<br> Individet har preferanser representert ved nyttefunksjonen: $U(x_1,x_2) = x_1x_2$

a) Hvor mye konsumeres av vare $x_1$ og $x_2$?
<br> b) Anta at prisen på vare $x_1$ reduseres til $p_1 = 4$
1) Hvor mye konsumeres nå av vare $x_1$ og $x_2$.
2) Illustrer substitusjon- og inntektseffekten i et mengde-diagram.

**Oppgave a**
<br> For å finne hvordan etterspørselen for vare $x_1$ og $x_2$ kan utrykkes må vi første finne to betingelser.
<br> Det er inntekten til individet som er gitt ved:
<br> $p_1x_1 + p_2x_2 = m$
<br> Og tangeringsbetingelsen som er gitt ved:
<br> $MSB = \frac{p_1}{p_2}$
<br> Vi vet også at formelen for MSB er gitt ved:
<br> $MSB = \frac{U'x_1}{U'x_2}$

Starter med å finne MSB:
<br> $MSB = \frac{U'x_1}{U'x_2} = \frac{x_2}{x_1}$
<br> Setter dette lik MSB og ganger med $x_1$ for å få $x_2$ alene:
<br> $\frac{x_2}{x_1} = \frac{p_1}{p_2}$
<br> $\frac{x_2}{x_1} *x_1 = \frac{p_1}{p_2} *x_1$
<br> $x_2 = \frac{p_1}{p_2} *x_1$

Setter utrykket for $x_2$ inn i likningen for inntekten til individet:
<br> $p_1x_1 + p_2*\frac{p_1}{p_2} *x_1 = m$
<br> $p_1x_1 + p_1x_1 = m$
<br> $2p_1x_1 = m$
<br> For å få $x_1$ helt alene kan vi dele på $2p_1$:
<br> $\frac{2p_1x_1}{2p_1} = \frac{m}{2p_1}$
<br> $x_1 = \frac{m}{2p_1}$

Setter inn utrykket for $x_1$ inn i utrykket for vare $x_2$:
<br> $x_2 = \frac{p_1}{p_2} * \frac{m}{2p_1}$
<br> $x_2 = \frac{m}{2p_2}$

Da har vi funnet utrykk for $x_1$ og $x_2$:
<br> $x_1 = \frac{m}{2p_1}$
<br> $x_2 = \frac{m}{2p_2}$

For å finne ut hvor mye som konsumeres av vare $x_1$ og $x_2$ kan vi nå sette inn verdiene for inntekt og pris:
<br> $x_1 = \frac{48}{2 * 6} = \frac{48}{12} = 4$
<br> $x_2 = \frac{48}{2 * 8} = \frac{48}{16} = 3$

Det betyr at konsumenten etterspør 4 enheter av vare $x_1$ og 3 enheter av vare $x_2$.

**Oppgave b (1)**
<br> Siden vi i forrige oppgave styrte bort fra å bruke faste priser og inntekt kan vi bruke det generelle utrykket vi fant isted til å beregne det nye konsumet:
<br> $x_1 = \frac{48}{2 * 4} = \frac{48}{8} = 6$
<br> $x_2 = \frac{48}{2 * 8} = \frac{48}{16} = 3$
<br> Det betyr at konsumenten etterspør 6 enheter av vare $x_1$ og 3 enheter av vare $x_2$.

**Oppgave b (2)**

Før prisreduskjon:
<br> $x_1 = \frac{48}{2 * 6} = \frac{48}{12} = 4$
<br> $x_2 = \frac{48}{2 * 8} = \frac{48}{16} = 3$

Etter prisreduksjon
<br> $x_1 = \frac{48}{2 * 4} = \frac{48}{8} = 6$
<br> $x_2 = \frac{48}{2 * 8} = \frac{48}{16} = 3$

```python
# Definerer den orginale budsjettkurven.
def bud_03(x_1):
    return 6 - x_1 * 0.75

# Definerer den nye budsjettkurven
def bud_04(x_1):
    return 6 - x_1 * 0.50
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("x_2")
ax.set_xlabel("x_1")

# Plotter grafer, linjer og tekst.
ax.plot(x_1, bud_03(x_1), color = "blue",   label = "Budsjettkurve (B_1)")
ax.plot(x_1, bud_04(x_1), color = "cyan",   label = "Budsjettkurve (B_2)")
ax.plot(4, 3,             color = "orange", label = "Optimal tilpasning (A)", marker = "o")
ax.plot(6, 3,             color = "purple", label = "Optimal tilpasning (B)", marker = "o")
ax.plot(5.3, 2.3,         color = "brown",  label = "Optimal tilpasning (C)", marker = "o")

# Begrenser aksene. 
ax.set(xlim = (0,14))
ax.set(ylim = (0,8))

# Gir figuren en tittel.
ax.set_title("Subtitusjon- og inntektseffekten")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser figuren.
plt.grid()
plt.show()
```
