## Marginal substitsusjonsbrøk

```python
# Laster inn pakker.
import numpy as np
from sympy import *
from matplotlib import pyplot as plt
```

<hr/>

#### Introduksjon til marginal substitusjonsbrøk (MSB)

*Hvor mye er jeg villig til å gi fra meg og fremdeles være fornøyd?*
<br> **MSB** forteller oss hvor mange enheter av et gode en konsument er villig til å avstå for å motta en enhet til av det andre gode, uten at nytten endres.
<br> **MSB** mellom to goder angir hvor mange enheter av det ene gode som akkurat kompenserer for at konsumenten frasier seg en enhet av det andre gode.
<br> **MSB** er ditt personlige bytteforhold mellom varene.
<br> Nedenfor skal vi se på en indifferenskurve.


```python
# Lager en funksjon for indifferenskurven.
def ind_0(x_0):
    return (5**(5/3))/x_0**(2/3)
```

```python
# Definerer ett intervall mellom 1, 15 med 10.000 punkter.
x_0 = np.linspace(1,15,10000)
```

```python
# Lager verdier for å illustrere i figuren.
ticks     = np.linspace(0,10,11)
orignal_x = np.linspace(0,5,6)
orginal_y = np.linspace(5,5,6)
fratatt_x = np.linspace(4,5,2)
fratatt_y = np.linspace(5,5,2)
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, ax = plt.subplots()

# Gir aksene navn.
ax.set_ylabel("Vare x_2")
ax.set_xlabel("Vare x_1")

# Plotter grafer, linjer og tekst.
ax.plot(x_0, ind_0(x_0),      color = "blue",   label = "Indifferenskurve (N = 5)")
ax.plot(orignal_x, orginal_y, color = "orange", label = "Orginal godekombinasjon",    linestyle = "--")
ax.plot(fratatt_x, fratatt_y, color = "red",    label = "Kompensasjon",               linestyle = "--")
ax.vlines(5, 0, 5,   color = "orange", linestyle = "--")
ax.vlines(4, 5, 5.8, color = "red",    linestyle = "--")
ax.plot(5,5,   color = "black", marker = "o")
ax.plot(4,5,   color = "black", marker = "o")
ax.plot(4,5.8, color = "black", marker = "o")
ax.text(5,5.2,    "A")
ax.text(4,6,      "B")
ax.text(3.3, 5.3, "MSB")

# Begrenser aksene. 
ax.set(xlim = (0,10))
ax.set(ylim = (0,10))

# Gir figuren en tittel.
ax.set_title("Indifferenskurve")

# Gir figuren en legend.
ax.legend(loc = "upper right")

# Viser alle verdier på x og y aksen.
plt.xticks(ticks)
plt.yticks(ticks)

# Viser figuren.
plt.grid()
plt.show()
```

På indifferenskurven over kan vi se at konsumenten orginalt fant seg i godekombinasjonen $x_1 = 5, x_2 = 5$ med en nytte på 250.
<br> På de røde kompensasjons linjene kan vi se at visst konsumenten blir fratatt 1 av gode $x_1$ trenger han ca 0.8 av gode $x_2$ for å være på samme nyttenivå.
<br> Det vil si at konsumenten er villig til å gi fra seg 1 av vare $x_1$ for 0.8 enhet ekstra av vare $x_2$.
<br> Avstanden fra punkt A til punkt B er **MSB**
<br>
<br> Vi kan se at bytteforholdet ikke er det samme hele veien, f.eks. når konsumenten har 10 av gode $x_1$ og ca. 3 av gode $x_2$, må vi gi fra oss ca. 3 av gode $x_1$ for å få 1 ekstra gode av $x_2$.
<br> Dette gir mening ettersom visst man har mye av det ene gode, vil man gi fra seg mer av det gode for å få 1 enhet mer av det andre gode.
<br> Stigningstallet til indifferenskurven forandrer seg hele tiden.

**MSB - Derivasjon**
<br> **MSB** tar naturligvis utgangspunkt i nyttefunksjonen til individet.
<br> Nedenfor skal vi se på et sjeldent tilfelle hvor nyttefunksjonen til individet er illustrert med en lineær indifferenskurve. Det forteller oss at godene er perfekte substituter og uansett hvor vi befinner oss på indifferenskurven så er konsumenten alltid villig til å gi fra seg nøyaktig den samme mengden av vare $x_2$ for å få 1 enhet mer av vare $x_1$.

```python
# Lager en funksjon for indifferenskurven.
def ind_1(x_1):
    return 10 - 2 * x_1
```

```python
# Definerer ett intervall mellom 0, 5 med 10 punkter.
x_1 = np.linspace(0, 5,100)
x_2 = np.linspace(1,15,100)
```

```python
# Lager subplot slik at jeg enkelt kan redigere det senere.
fig, axs = plt.subplots(1,2,sharey=True, figsize=(20,5))

# Gir aksene navn.
axs[0].set_ylabel("Vare x_2")
axs[0].set_xlabel("Vare x_1")
axs[1].set_ylabel("Vare x_2")
axs[1].set_xlabel("Vare x_1")

# Plotter grafer, linjer og tekst.
axs[0].plot(x_1, ind_1(x_1), color = "blue", label = "Indifferenskurve (N = 10)")
axs[1].plot(x_2, ind_0(x_2),      color = "blue",   label = "Indifferenskurve (N = 5)")

# Begrenser aksene. 
axs[0].set(xlim = (0,10))
axs[0].set(ylim = (0,10))
axs[1].set(xlim = (0,10))
axs[1].set(ylim = (0,10))

# Gir figuren en tittel.
axs[0].set_title("Perfekt substituter (Lineær)")
axs[1].set_title("Standartilfelle (Krummet konveks kurve)")

# Gir figuren en legend.
axs[0].legend(loc = "upper right")
axs[1].legend(loc = "upper right")

# Viser figuren.
axs[0].grid()
axs[1].grid()
plt.show()
```

Fra funksjonen til indifferenskurven ($y = 10 - 2*x$) kan vi se at konsumenten er alltid villig til å gi fra seg 2 enheter av vare $x_2$ for 1 enhet ekstra av vare $x_1$.
<br> Vi kan derfor si at **MSB** er gitt ved tallverdien på hellningen på indifferenskurven.
<br> På det tidligere eksempelet er det ikke like enkelt, der endrer stigningstallet seg hele tiden. Vi kan se i starten av kurven er konsumenten villig til å gi fra seg mange enheter av vare $x_2$ for 1 enhet ekstra av vare $x_1$. Lenger nede på indifferenskurven ser vi at kurven faller mindre, som betyr at konsumenten er villig til å gi fra seg mindre enheter av vare $x_2$ for 1 enehet av vare $x_1$.
<br> For å finne **MSB** i dette eksempelet bruker vi derivasjon. $Derivasjon = Stigningstallet = MSB$

**Eksempel: Beregne MSB**
<br> $U(x_1,x_2) = x_1x_2$
<br> Siden vi har to goder må vi partiell derivere.
<br> **MSB** for en funksjon med to variabler finner vi ved:
<br> $MSB = \frac{grensenytte vare x_1}{grensenytte vare x_2}$
<br> Grensenytte av vare $x_1$ finner vi ved å partiell derivere nyttefunksjonen $U$ med hensyn på $x_1$. $=U'x_1$
<br> Grensenytte av vare $x_2$ finner vi ved å partiell derivere nyttefunksjonen $U$ med hensyn på $x_2$. $=U'x_2$
<br> $MSB = \frac{U'x_1}{U'x_2}$

#### Beregne MSB (Oppgaver)

**Oppgave 1)**
<br> Vi ser på en konsument med følgende nyttefunksjon:
<br> $U(x_1,x_2) = 4x_1 + 2x_2^{\frac{1}{2}}$
<br> Finn utrykket for MSB for denne personen.
<br>
<br> Fra eksempelet lenger opp vet vi at MSB er gitt ved: $MSB = \frac{grensenytte vare x_1}{grensenytte vare x_2} = \frac{U'x_1}{U'x_2}$
<br> Vi starter med å parsiellderivere med hensyn på $x_1$.
<br> $U'x_1 = 4x_1 + 2x_2^{\frac{1}{2}}$
<br> $U'x_1 = 4x_1^{1}$
<br> $U'x_1 = 4*1*x_1^{1-1}$
<br> $U'x_1 = 4*1*x_1^{0}$
<br> $U'x_1 = 4*1*1$
<br> $U'x_1 = 4$
<br> Deretter parsiellderivere vi med hensyn på $x_2$.
<br> $U'x_2 = 4x_1 + 2x_2^{\frac{1}{2}}$
<br> $U'x_2 = 2x_2^{\frac{1}{2}}$
<br> $U'x_2 = 2*\frac{1}{2}*x_2^{\frac{1}{2} - 1}$
<br> $U'x_2 = 2*\frac{1}{2}*x_2^{-\frac{1}{2}}$
<br> $U'x_2 = 1*x_2^{-\frac{1}{2}}$
<br> $U'x_2 = x_2^{-\frac{1}{2}}$
<br> Vi har nå partiellderiverte både $x_1$ og $x_2$.
<br> Så setter vi utrykkene sammen i formelen for MSB.
<br> $MSB = \frac{U'x_1}{U'x_2} = \frac{4}{x_2^{-\frac{1}{2}}} = 4x_2^{\frac{1}{2}}$

**Oppgave 2)**
<br> Vi ser på en konsument med følgende nyttefunksjon:
<br> $U(x_1,x_2) = x_1^{0.1}x_2^{0.2}$
<br> Finn utrykket for MSB for denne personen.
<br> Vi skal se på 2 måter å løse denne oppgaven på, partiellderivasjon og en "snarvei".
<br>
<br> *Løsning 1 - Partiellderivasjon*
<br> Fra eksempelet lenger opp vet vi at MSB er gitt ved: $MSB = \frac{grensenytte vare x_1}{grensenytte vare x_2} = \frac{U'x_1}{U'x_2}$
<br> Vi starter med å parsiellderivere med hensyn på $x_1$.
<br> $U'x_1 = x_1^{0.1}*x_2^{0.2}$
<br> $U'x_1 = 0.1 * x_1^{0.1-1}*x_2^{0.2}$
<br> $U'x_1 = 0.1x_1^{-0.9}*x_2^{0.2}$
<br> Deretter parsiellderivere vi med hensyn på $x_2$.
<br> $U'x_2 = x_1^{0.1}*x_2^{0.2}$
<br> $U'x_2 = x_1^{0.1}*0.2 * x_2^{0.2-1}$
<br> $U'x_2 = x_1^{0.1}*0.2x_2^{-0.8}$
<br> Vi har nå partiellderiverte både $x_1$ og $x_2$.
<br> Så setter vi utrykkene sammen i formelen for MSB.
<br> $MSB = \frac{U'x_1}{U'x_2} = \frac{0.1x_1^{-0.9}*x_2^{0.2}}{x_1^{0.1}*0.2x_2^{-0.8}} = \frac{0.1 * x_2^{0.2} * x_1^{-0.9}}{0.2 * x_1^{0.1} * x_2^{-0.8}} = \frac{0.1 * x_2^{0.2} * x_2^{0.8}}{0.2 * x_1^{0.1} * x_1{0.9}} = \frac{0.1x_2^{1}}{0.2x_1^{1}} = \frac{0.1x_2}{0.2x_1} = \frac{1x_2}{2x_1} = \frac{x_2}{2x_1}$
<br> 
<br> *Løsning 2 - Snarvei*
<br> $U(x_1,x_2) = x_1^{0.1}x_2^{0.2} = x_1^{a}x_2^{b}$
<br> $MSB = \frac{a}{b} * \frac{x_2}{x_1} = \frac{0.1}{0.2} * \frac{x_2}{x_1} = \frac{1}{2} * \frac{x_2}{x_1} = \frac{1 * x_2}{2 * x_1} = \frac{x_2}{2x_1}$

**Oppgave 3)**
<br> Vi ser på en konsument med følgende nyttefunksjon:
<br> $U(x,y) = 5xy+5x$
<br> Finn utrykket for MSB for denne personen.
<br>
<br> Fra eksempelet lenger opp vet vi at MSB er gitt ved: $MSB = \frac{grensenytte vare x}{grensenytte vare y} = \frac{U'x}{U'y}$
<br> Vi starter med å parsiellderivere med hensyn på $x$.
<br> $U'x = 5xy+5x$
<br> $U'x = 5*1*x^{1-1}*y+5*1x^{1-1}$
<br> $U'x = 5*1*1*y+5*1*1$
<br> $U'x = 5y+5$
<br> Deretter parsiellderivere vi med hensyn på $y$.
<br> $U'y = 5xy+5x$
<br> $U'y = 5*x*1*y^{1-1}$
<br> $U'y = 5*x*1*1$
<br> $U'y = 5x$
<br> Vi har nå partiellderiverte både $x$ og $y$.
<br> Så setter vi utrykkene sammen i formelen for MSB.
<br> $MSB = \frac{5y+5}{5x} = \frac{y+1}{x}$
